위의 식은 Gradient Descent 를 구하는 과정이다.
위의 식을 pytorch 로 고치면 아래와 같다.

```python
    # 위의 a(알파) 는 learning_rate 이다
    W = W - learning_rate * torch.mean((hypothesis - y_data) * x_data)

```

````python
import torch
import math

# Hypothesis function H(x) 정의

x_data = torch.FloatTensor([[1], [2], [3]])
y_data = torch.FloatTensor([[1], [2], [3]])

W = torch.zeros(1, requires_grad=True)

for step in range(2001):

    # Our hypothesis for linear model X * W
    hypothesis = x_data * W

    # cost / loss function
    cost = torch.mean((hypothesis - y_data) * x_data)

    # Gradient Descent 의 구현
    # Minimize: Gradient Descent using derivative: W -= learning_rate * derivative
    learning_rate = 0.01
    W = W - learning_rate * cost

    if step % 100 == 0:
        print(step, cost.item(), W.item())
````
