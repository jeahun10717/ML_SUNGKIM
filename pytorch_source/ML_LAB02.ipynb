````python
import torch
import math
````

````python
# Hypothesis function H(x) 정의

# x_train = torch.tensor([1, 2, 3])
# y_train = torch.tensor([1, 2, 3])

x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])

print(x_train.shape)

W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
````

````python
# Gradient Descent
optimizer = torch.optim.SGD([W, b], lr = 0.01)
````

````python
for step in range(2001):

    # H(x) 정의
    hypothesis = x_train * W + b

    # cost function 정의
    cost = torch.mean((hypothesis - y_train) ** 2)

    # cost로 H(x) 개선
    # gradient를 0으로 초기화
    optimizer.zero_grad() 
    # 비용 함수를 미분하여 gradient 계산
    cost.backward() 
    # W와 b를 업데이트
    optimizer.step() 

    # 100번마다 로그 출력
    if step % 100 == 0:
        print(step, cost.item(), W.item(), b.item())
````

위의 소스에서 optimizer.zero_grad() 가 필요한 이유

``` python
import torch
w = torch.tensor(2.0, requires_grad=True)

nb_epochs = 20
for epoch in range(nb_epochs + 1):

  z = 2*w

  z.backward()
  print('수식을 w로 미분한 값 : {}'.format(w.grad))
```

위의 식을 실행하면 `w == 2, 4, 6, 8 ...` 처럼 지속적으로 증가한다. 이는 pytorch 의 특성으로 `backward` 를 통해 구한 미분값을 누적시킨다. 따라서 그 값을 매번 `zero_grad()` 로 초기화 해 주어야 한다. 아래 소스 참조

````python
# zero_grad() 안쓴 소스
import torch
w = torch.tensor(2.0, requires_grad=True)

optimizer = torch.optim.SGD([w], lr = 0.01)
nb_epochs = 20
for epoch in range(nb_epochs + 1):

  z = 2*w
  # optimizer.zero_grad()
  z.backward()
  print('수식을 w로 미분한 값 : {}'.format(w.grad))
````

````python
# zero_grad() 쓴 소스
import torch
w = torch.tensor(2.0, requires_grad=True)

optimizer = torch.optim.SGD([w], lr = 0.01)
nb_epochs = 20
for epoch in range(nb_epochs + 1):

  z = 2*w
  optimizer.zero_grad()
  z.backward()
  print('수식을 w로 미분한 값 : {}'.format(w.grad))
````
